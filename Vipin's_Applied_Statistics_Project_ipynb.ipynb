## Checking and Handling Outliers##

```python
# Importing necessary libraries for visualizations.
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Key_variables = ['Income','MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']

# Creating box plots for the key variables.
for variable in Key_variables:
    plt.figure(figsize=(6, 3))
    sns.boxplot(y=df[variable])
    plt.title(f'Box Plot of {variable}')
    plt.ylabel(variable)
    plt.show()
```

```python
# Handling Outliers.
def remove_outliers(data, columns):
    for column in columns:
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return data

Key_variables = ['Income','MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']
```

```python
# Setting new variable for future use.
new_df = remove_outliers(df, Key_variables)
```

```python
# Again checking for outliers after handling.
Key_variables = ['Income','MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']

for variable in Key_variables:
    plt.figure(figsize=(4, 4))
    sns.boxplot(y=new_df[variable])
    plt.title(f'Box Plot of {variable} (Outliers Removed)')
    plt.ylabel(variable)
    plt.show()
```

## Task 3 - Probability Distributions üé≤

- **Identify variables that could follow specific probability distributions (e.g., Binomial, Normal). Calculate probabilities and expected values based on these distributions.**



#### Here the columns present in the Key_variables(taken by me) don't have data necessary for binomial distribution. So we are cheking for normal distribution only. ####

```python
# Checking for Normal Distribution:
Key_variables = ['Income','MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']

# Iterating through each variable and creating a histogram.
for variable in Key_variables:
    plt.figure(figsize=(6, 4))
    sns.histplot(new_df[variable], kde=True)
    plt.title(f'Histogram of {variable}')
    plt.xlabel(variable)
    plt.ylabel('Frequency')
    plt.show()
```

#### By making histogram, we came to know that only 'Income' column shows normal distribution; Hence we are calculating probability distribution and expected value only for 'Income' column. ####

```python
mean = new_df['Income'].mean()
std = new_df['Income'].std()

# Probability of income less than or equal to 45000
prob_less_than_45000 = stats.norm.cdf(45000, mean, std)
print(f"Probability of Income <= 45000: {prob_less_than_45000}")

# Probability of income grater than 45000
prob_more_than_45000 = 1 - prob_less_than_45000
print(f"Probability of Income > 45000: {prob_more_than_45000}")

# Expected income
expected_income = mean
print(f"Expected Income: {expected_income}")
```

## Task 4: Insights and Customer Segmentation üìà

```python
# Calculating the correlation matrix for all numerical columns.
numerical_df = new_df.select_dtypes(include=np.number)
correlation_matrix = numerical_df.corr().round(2)
correlation_matrix
```
